```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key, it in ax.spines.items():    
        it.set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering', 'Computer Science']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

f, axarr = plt.subplots(6, 3, figsize=(9, 18))

for sp in range(0,6):
    axarr[sp, 0].plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    axarr[sp, 0].plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key, it in axarr[sp, 0].spines.items():    
        it.set_visible(False)
    axarr[sp, 0].set_xlim(1968, 2011)
    axarr[sp, 0].set_ylim(0,100)
    axarr[sp, 0].set_yticks([0,100])
    axarr[sp, 0].set_title(stem_cats[sp])
    axarr[sp, 0].axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    axarr[sp, 0].tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if sp == 0:
        axarr[sp, 0].text(2005, 87, 'Men')
        axarr[sp, 0].text(2002, 8, 'Women')
    elif sp == 5:
        axarr[sp, 0].text(2005, 62, 'Men')
        axarr[sp, 0].text(2001, 35, 'Women')
        axarr[sp, 0].tick_params(labelbottom='on')
        
for sp in range(0,5):
    axarr[sp, 1].plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    axarr[sp, 1].plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key, it in axarr[sp, 1].spines.items():    
        it.set_visible(False)
    axarr[sp, 1].set_xlim(1968, 2011)
    axarr[sp, 1].set_ylim(0,100)
    axarr[sp, 1].set_yticks([0,100])
    axarr[sp, 1].axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    axarr[sp, 1].set_title(lib_arts_cats[sp])
    axarr[sp, 1].tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if sp == 0:
        axarr[sp, 1].text(2005, 87, 'Men')
        axarr[sp, 1].text(2002, 8, 'Women')
    elif sp == 4:
        axarr[sp, 1].text(2005, 62, 'Men')
        axarr[sp, 1].text(2001, 35, 'Women')
        axarr[sp, 1].tick_params(labelbottom='on')

for sp in range(0,6):
    axarr[sp, 2].plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    axarr[sp, 2].plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    for key, it in axarr[sp, 2].spines.items():    
        it.set_visible(False)
    axarr[sp, 2].set_xlim(1968, 2011)
    axarr[sp, 2].set_ylim(0,100)
    axarr[sp, 2].set_yticks([0,100])
    axarr[sp, 2].axhline(50, c=(171/255, 171/255, 171/255), alpha=0.3)
    axarr[sp, 2].set_title(other_cats[sp])
    axarr[sp, 2].tick_params(bottom="off", top="off", left="off", right="off", labelbottom='off')
    if sp == 0:
        axarr[sp, 2].text(2005, 87, 'Men')
        axarr[sp, 2].text(2002, 8, 'Women')
    elif sp == 5:
        axarr[sp, 2].text(2005, 62, 'Men')
        axarr[sp, 2].text(2001, 35, 'Women')  
        axarr[sp, 2].tick_params(labelbottom='on')
plt.plot(women_degrees['Year'], women_degrees['Biology'])
plt.savefig('biology_degrees.png')
plt.show()
```
